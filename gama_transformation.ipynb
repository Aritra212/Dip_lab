### Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read Image Function

```python
def Read_img(path):
    img= open(path,"r")
    g= img.readline()
    c= img.readline()
    rc= img.readline()
    max= int(img.readline())
    
    [col,row]=[int (i) for i in rc.split()]
    k=np.zeros((row,col),np.int64)  #Arry to store all image pixels
    for i in range (row):
        for j in range(col):
            k[i,j]=img.readline()
    
    img.close()
    plt.imshow(k,cmap='gray')
    return k,g,c,rc,max,col,row
```

### Store/Write O/p Image Function

```python
def Write_img(path,k,g,c,rc,max,col,row):
    img= open(path,"w")
    img.write(g)
    img.write(c)
    img.write(rc)
    img.write(str(max))
    
    for i in range(row):
        for j in range(col):
            img.write("%d\n"%k[i,j])
    img.close()
```

### Gama or Power Law Transformation Function

```python
def gama_transform(k,max,row,col,a,g):
    for i in range(row):
        for j in range(col):
            k[i,j]=a*(k[i,j]**g)
    plt.imshow(k,cmap='gray')
```

### Main function

```python
if __name__ == '__main__':
    k1,g1,c1,rc1,max1,col1,row1= Read_img("..\images\DIP3E_Original_Images_CH03\Fig0320(2)(2nd_from_top).pgm")
    a=1
    g=0.5
    gama_transform(k1,max1,row1,col1,a,g)
    Write_img("outputs\\neg_img.pgm",k1,g1,c1,rc1,max1,col1,row1)
```
